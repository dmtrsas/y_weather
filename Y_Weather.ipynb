```python
import requests
import csv
from datetime import datetime

locations = {"Msk": ["55.7512", "37.6184"], "Spb": ["59.9375", "30.3086"], "Kaz": ["55.7963", "49.1088"],
             "Tul": ["54.2048", "37.6184"], "Nsb": ["55.0188", "82.9339"]}  # словарь с координатами нужных городов
header_key = {'X-Yandex-API-Key': '3288cf45-30c6-4dd0-8425-778dabf102ce'}  # ключ API

with open('Weather_forecast_result.csv', 'w', newline='') as wfr:
    writer = csv.writer(wfr)
    header = ['city', 'date', 'hour', 'temperature_c', 'pressure_mm', 'is_rainy']
    writer.writerow(header)  # записываем хэдер

    for loc in locations:
        rsp = requests.get(
            f"https://api.weather.yandex.ru/v2/forecast?lat={locations[loc][0]}&lon={locations[loc][1]}&lang=ru_RU&limit=7&hours=true&extra=true",
            headers=header_key).json()  # для каждой локации обращаемся к API и получаем ответ в виде JSON

        city = rsp['geo_object']['locality']['name']  # находим название города
        for dt in rsp['forecasts']:  # для каждой даты в прогнозе
            if dt['hours']: # если есть разбивка по часам - обрабатываем
                for hour in dt['hours']:  # у которой есть прогноз по часам - ограничение только на 96 часов
                    res_row.append(city)  # добавляем город,
                    res_row.append(datetime.strptime(dt['date'],"%Y-%m-%d").strftime("%d.%m.%Y"))  # добавляем дату в нужном формате,
                    res_row.append(hour['hour'])  # добавляем час в дате,
                    res_row.append(hour['temp'])  # добавляем температуру в часе,
                    res_row.append(hour['pressure_mm'])  # добавляем давление в часе
                    if hour['prec_type'] == 1 and hour['prec_strength'] > 0:  # если тип осадков == дождь и сила > 0
                        res_row.append(1)  # то флаг = 1
                    else:
                        res_row.append(0)  # иначе - флаг = 0
                    writer.writerow(res_row)  # пишем строку в файл
                    res_row = []
            else:           # если нет разбивки по часам - не обрабатываем
                continue
```
